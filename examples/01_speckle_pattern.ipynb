# Simulating a speckle pattern with shot noise

Import basic routines

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
```

Import speckle related routines

```python
sys.path.insert(0, '../scripts/')
from speckle_routines import *
```

parameters for the speckle pattern

```python
nx,ny = 100,120 # size of array x,y
kbar = 0.1   # mean photon density (photons per pixel)
modes = 1     # number of modes in the speckle pattern
specklesize=5 # speckle size in pixel units
```

Varying the speckle size

```python
specklesize=2
speckle1 = model_speckles_modes([nx,ny],specklesize,modes)
specklesize=5
speckle2 = model_speckles_modes([nx,ny],specklesize,modes)
specklesize=10
speckle3 = model_speckles_modes([nx,ny],specklesize,modes)

plt.figure()

plt.subplot(1,3,1)
plt.imshow(speckle1,interpolation='none')

plt.subplot(1,3,2)
plt.imshow(speckle2,interpolation='none')

plt.subplot(1,3,3)
plt.imshow(speckle3,interpolation='none')

plt.show()
```

Varying number of modes

```python
specklesize=5
nmodes = [1,10,1000]

plt.figure()

for i in range(len(nmodes)):
    plt.subplot(1,len(nmodes),i+1)
    speckle = model_speckles_modes([nx,ny],specklesize,nmodes[i])
    plt.imshow(speckle,interpolation='none')
    plt.tight_layout()
plt.show()
```

Varying the photon density (number of photons/pixel)

```python
specklesize=5
modes = 1
nkbar = [1,0.1,0.01] 

plt.figure()

for i in range(len(nmodes)):
    plt.subplot(1,len(nmodes),i+1)
    speckle = simulate_speckles_with_shot_noise([nx,ny],modes,specklesize,nkbar[i])
    plt.imshow(speckle,interpolation='none')
    plt.tight_layout()
plt.show()

speckle = simulate_speckles_with_shot_noise([nx,ny],modes,specklesize,kbar)
```
